```python
classes = ['apple_pie', 'sushi', 'falafel', 'cannoli', 'ramen']  # example, must match exactly
```

```python
class FilteredFoodDataset(Dataset):
    def __init__(self, dataset, selected_classes):
        self.dataset = dataset
        self.selected_classes = selected_classes
        self.class_to_idx = {cls_name: i for i, cls_name in enumerate(selected_classes)}
        self.filtered_data = []

        for img, label in self.dataset:
            cls_name = self.dataset.classes[label]
            if cls_name in selected_classes:
                new_label = self.class_to_idx[cls_name]  # remap 0-4
                self.filtered_data.append((img, new_label))

    def __len__(self):
        return len(self.filtered_data)

    def __getitem__(self, idx):
        return self.filtered_data[idx]
```

```python
from torchvision import datasets, transforms

transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406],
                         [0.229, 0.224, 0.225])
])

full_dataset = datasets.ImageFolder('/content/food101/data/food-101-tiny/train', transform=transform)
dataset = FilteredFoodDataset(full_dataset, classes)
print("Total samples after filtering:", len(dataset))
```

```python
from torch.utils.data import random_split, DataLoader

train_size = int(0.8 * len(dataset))
valid_size = len(dataset) - train_size
train_dataset, valid_dataset = random_split(dataset, [train_size, valid_size])

train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
valid_loader = DataLoader(valid_dataset, batch_size=32, shuffle=False)

print("Train samples:", train_size, "Valid samples:", valid_size)
```

```python
from torchvision import models
import torch.nn as nn

# Modeli yüklə
model = models.resnet18(weights=models.ResNet18_Weights.IMAGENET1K_V1)

# Bütün qatları freeze et
for param in model.parameters():
    param.requires_grad = False

# Son fully connected layeri 5 sinif üçün dəyiş
model.fc = nn.Linear(512, 5)

# GPU varsa göndər
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = model.to(device)

print(model)
```

```python
import torch.optim as optim
import torch

criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.fc.parameters(), lr=0.0001)
```

```python
def train_model(model, train_loader, valid_loader, criterion, optimizer, epochs=10):
    train_acc_history = []
    valid_acc_history = []

    for epoch in range(epochs):
        model.train()
        running_corrects = 0
        for images, labels in train_loader:
            images, labels = images.to(device), labels.to(device)

            optimizer.zero_grad()
            outputs = model(images)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()

            _, preds = torch.max(outputs, 1)
            running_corrects += torch.sum(preds == labels.data)

        epoch_train_acc = running_corrects.double() / len(train_loader.dataset)
        train_acc_history.append(epoch_train_acc.item())

        # Validation
        model.eval()
        valid_corrects = 0
        with torch.no_grad():
            for images, labels in valid_loader:
                images, labels = images.to(device), labels.to(device)
                outputs = model(images)
                _, preds = torch.max(outputs, 1)
                valid_corrects += torch.sum(preds == labels.data)

        epoch_valid_acc = valid_corrects.double() / len(valid_loader.dataset)
        valid_acc_history.append(epoch_valid_acc.item())

        print(f"Epoch {epoch+1}/{epochs} | Train Acc: {epoch_train_acc:.4f} | Valid Acc: {epoch_valid_acc:.4f}")

    return train_acc_history, valid_acc_history
```

```python
v1_train_acc, v1_valid_acc = train_model(model, train_loader, valid_loader, criterion, optimizer)
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

model.eval()
all_preds = []
all_labels = []

with torch.no_grad():
    for images, labels in valid_loader:
        images, labels = images.to(device), labels.to(device)
        outputs = model(images)
        _, preds = torch.max(outputs, 1)
        all_preds.extend(preds.cpu().numpy())
        all_labels.extend(labels.cpu().numpy())

cm = confusion_matrix(all_labels, all_preds)
plt.figure(figsize=(6,5))
sns.heatmap(cm, annot=True, fmt="d", xticklabels=classes, yticklabels=classes)
plt.xlabel('Predicted')
plt.ylabel('True')
plt.show()
```
